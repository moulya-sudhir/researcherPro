```python
import pandas as pd
import json
```

```python
df1 = pd.read_csv('data/csvs/2023acl-long_article_data.csv')
df2 = pd.read_csv('data/csvs/2023conll-1_article_data.csv')
df3 = pd.read_csv('data/csvs/2023emnlp-main_article_data.csv')
```

```python
data = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
data.to_csv('data/all_articles.csv', index=False)
```

```python
data.info()
```

```python
from glob import glob
all_json_paths = glob("data/**/*.json", recursive=True)
```

```python
all_jsons = {}

for json_path in all_json_paths:
    # Load JSON file
    with open(json_path, 'r') as f:
        json_file = json.load(f)
    
    # Extract title
    title = json_file['title']
    
    # Delete title from the json_file keys
    del json_file['title']

    # Add it to all_jsons
    all_jsons[title] = json_file
```

```python
with open('data/all_jsons.json','w') as f:
    json.dump(all_jsons, f, indent=4)
```
